## 1. Data Loading and Inspection

```python
# Cell 1: Import Libraries and Load Data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Settings for better visualizations
sns.set_style("whitegrid")

# Load the dataset
df = pd.read_csv('Leads.csv')

# Display the first few rows
print("First 5 rows of the dataset:")
display(df.head())

# Get a concise summary of the dataframe
print("\nDataset Info:")
df.info()

# Get summary statistics for numerical columns
print("\nSummary Statistics:")
display(df.describe())
```

## 3. Modeling

## 3.1 Preprocessing

```python
# Cell 5: Data Cleaning and Preprocessing
from sklearn.model_selection import train_test_split

# --- Data Cleaning ---

# For this sprint, we'll drop columns that are not immediately useful, are redundant, or have too many unique values to encode easily.
cols_to_drop = ['Prospect ID', 'Lead Quality', 'Asymmetrique Activity Index', 
                'Asymmetrique Profile Index', 'Asymmetrique Activity Score', 'Asymmetrique Profile Score',
                'Tags', 'Last Notable Activity', 'Last Activity', 'Lead Profile']

df_clean = df.drop(cols_to_drop, axis=1)

# For speed, we will drop rows with any remaining missing values.
# A more advanced approach would be to impute them, but this is faster for our sprint.
df_clean.dropna(inplace=True)

# --- Preprocessing ---

# Convert all categorical columns to numerical using one-hot encoding
categorical_cols = df_clean.select_dtypes(include=['object']).columns
df_final = pd.get_dummies(df_clean, columns=categorical_cols, drop_first=True)

# Separate features (X) and the target variable (y)
X = df_final.drop('Converted', axis=1)
y = df_final['Converted']

# Split data into a training set (70%) and a testing set (30%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

print("Shape of training data (X_train):", X_train.shape)
print("Shape of testing data (X_test):", X_test.shape)
```

## 3.2 Model Training

```python
# Cell 6: Model Building and Training
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

# Initialize the models with settings for reproducibility
log_reg = LogisticRegression(max_iter=1000, random_state=42)
rand_forest = RandomForestClassifier(random_state=42)
xgb = XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='logloss')

# Train the models on the training data
print("Training Logistic Regression...")
log_reg.fit(X_train, y_train)

print("Training Random Forest...")
rand_forest.fit(X_train, y_train)

print("Training XGBoost...")
xgb.fit(X_train, y_train)

print("\nAll models have been trained successfully! âœ…")
```

## 3.3 Evaluation

```python
# Cell 7: Model Evaluation
from sklearn.metrics import roc_auc_score, accuracy_score, precision_score

# --- Make Predictions ---
# Get class predictions (0 or 1)
y_pred_log_reg = log_reg.predict(X_test)
y_pred_rand_forest = rand_forest.predict(X_test)
y_pred_xgb = xgb.predict(X_test)

# Get prediction probabilities (needed for AUC-ROC)
y_prob_log_reg = log_reg.predict_proba(X_test)[:, 1]
y_prob_rand_forest = rand_forest.predict_proba(X_test)[:, 1]
y_prob_xgb = xgb.predict_proba(X_test)[:, 1]

# --- Calculate Metrics ---
metrics = {
    'Model': ['Logistic Regression', 'Random Forest', 'XGBoost'],
    'AUC-ROC Score': [roc_auc_score(y_test, y_prob_log_reg), 
                      roc_auc_score(y_test, y_prob_rand_forest), 
                      roc_auc_score(y_test, y_prob_xgb)],
    'Accuracy': [accuracy_score(y_test, y_pred_log_reg), 
                 accuracy_score(y_test, y_pred_rand_forest), 
                 accuracy_score(y_test, y_pred_xgb)],
    'Precision': [precision_score(y_test, y_pred_log_reg), 
                  precision_score(y_test, y_pred_rand_forest), 
                  precision_score(y_test, y_pred_xgb)]
}

results_df = pd.DataFrame(metrics).round(3) # Rounding to 3 decimal places
results_df.set_index('Model', inplace=True) # Set Model as the index

print("Model Performance Comparison:")
display(results_df)
```
